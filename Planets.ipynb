# PLANETS DATASETS

* Seaborn's planets dataset contains many different information about planets' 
orbits. This data set includes various features such as year, planet name,
progression method, degrees, clarity. Many different analyzes can be made o 
this data. Here's some analyis:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = sns.load_dataset("planets")
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
#If the variables we set as categories have object dtype, the category is 
#We can change it to
df["method"] = pd.Categorical(df["method"])
```

```python
df.info()
```

```python
#Checking how many different methods are available
df["method"].value_counts()
```

```python
df["method"].unique()
```

```python
df["method"].nunique()
```

```python
#How many NaN values ​​are there?
df["distance"].value_counts(dropna = False)
```

```python
# Number of planets discovered by year
planets_per_year = df.groupby('year')['method'].count()

plt.figure(figsize=(10, 6))
plt.plot(planets_per_year.index, planets_per_year.values, marker='o')
plt.title('Number of Planets Discovered Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Planets')
plt.grid(True)
plt.show()
```

```python
#By looking at the discovery methods used to find planets, we can see which methods are more common.
method_counts = df["method"].value_counts()

plt.figure(figsize= (8,5))
method_counts.plot( kind='barh' , color = 'skyblue')
plt.title("Number of Planets Discovered by Method")
plt.xlabel("Number of Planets")
plt.ylabel("Discovery method")
plt.show()
```

```python
# Status of mass data (non-NaN values)
plt.figure(figsize=(8,5))
sns.histplot(df['mass'].dropna(), bins=30, kde = True, color = "yellow")
plt.title("Distribution of Planet Masses")
plt.xlabel("Mass")
plt.ylabel("Frequency")
plt.show()
```

```python
#Distance distribution of planets
plt.figure(figsize=(8,5))
sns.histplot(df["distance"].dropna(), bins = 30, kde = True, color = "red")
plt.title("Distribution of Planet Distances")
plt.xlabel("Distance (light years)")
plt.ylabel("Frequency")
plt.show()
```
